```python
import tweetnlp
import pandas as pd
import re

#'sentencepiece '
model = tweetnlp.load_model('sentiment')
```

```python
def predict_sentiment(tweet: str):
    return model.sentiment(tweet)['label']
def clean_string(string):
    return re.sub(r"[,\[\]']", '', string)
```

```python
df = pd.read_csv("../../data/data_clean.csv",index_col=0)
```

```python
df['tweet_cleaned_joined'] = df['tweet_cleaned'].apply(clean_string)
```

```python
df['sentiment'] = df['tweet_cleaned_joined'].apply(predict_sentiment)
```
